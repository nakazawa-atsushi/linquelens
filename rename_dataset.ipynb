```python
# rename filename from number to focus value
import os
import sys, csv
import shutil
```

```python
path = "data/202409181754/"
path_new = "data/202409181754-new"

#os.mkdir(path_new)

with open(os.path.join(path,"data.csv"), "rt") as f:
    reader = csv.reader(f)
    data = [row for row in reader]

for val, fn in data:
    fn = "data/" + fn
    
    print("copy ", fn, os.path.join(path_new,val + ".bmp"))
    shutil.copy(fn, os.path.join(path_new,val + ".bmp"))
```
